```python
# BTS Streaming Analysis
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "song": ["Dynamite", "Butter", "DNA", "Fake Love", "Boy With Luv"],
    "year": [2020, 2021, 2017, 2018, 2019],
    "spotify_popularity": [95, 98, 85, 88, 92],
    "billboard_peak": [1, 1, 67, 10, 8]
})

df
```

```python
plt.figure(figsize=(10,5))

plt.bar(df["song"], df["spotify_popularity"])

plt.title("BTS Spotify Popularity")
plt.xlabel("Songs")
plt.ylabel("Spotify Popularity")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.bar(df["song"], df["billboard_peak"])

plt.title("BTS Billboard Peak Ranking")
plt.xlabel("Songs")
plt.ylabel("Peak Billboard Rank")

plt.gca().invert_yaxis()

plt.show()
```
